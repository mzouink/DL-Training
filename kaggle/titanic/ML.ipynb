```python
import pandas as pd
import numpy as np
```

```python
import tensorflow as tf
```

```python
tf.config.list_physical_devices('GPU')
```

```python
train_df = pd.read_csv("data/formatted_train.csv")
test_df = pd.read_csv("data/formatted_test.csv")
```

```python
print(train_df.shape)
print("train= ",train_df.info())
train_df.head()
```

```python
print(test_df.shape)
print("train= ",test_df.info())
test_df.head()
```

```python
def getXY(data):
    x = np.array(data,dtype="float32")
    x =  np.delete(x,0,1)
    y = np.array(data.loc[:,"Survived"],dtype="float32")
    return x,y

X,Y = getXY(train_df)
# X_test,Y_test = getXY(test_df)
print(X.shape)
print(Y.shape)
```

```python
model = tf.keras.models.Sequential(
[
    tf.keras.layers.Dense(64, activation='relu'),
#     tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(64, activation='relu'),
#     tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(64, activation='relu'),
#     tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(64, activation='relu'),
#     tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(64, activation='relu'),
#     tf.keras.layers.Dropout(0.2),
#     tf.keras.layers.Dense(64, activation='relu'),
#     tf.keras.layers.Dropout(0.2),
#     tf.keras.layers.Dense(64, activation='relu'),
#     tf.keras.layers.Dropout(0.2),
#     tf.keras.layers.Dense(64, activation='relu'),
#     tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(1, activation='sigmoid')
])
```

```python
model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
```

```python
model.fit(X,Y, epochs=500)
```
